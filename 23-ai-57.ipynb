```
import pandas as pd
```

```
#Load the dataset

df = pd.read_csv('passenger.csv')
```

```
df
```

```
#total number of passengers

df['passenger_id'].count()
```

```
#how many passengers are traveling to each destination

df.groupby('destination')['passenger_id'].count()
```

```
#average ticket price paid by male and female passengers

df.groupby('gender')['ticket_price'].mean('ticket_price')
```

```
#function that returns a list of names of passengers going to the specified destination

def get_passengers_by_destination(dest):
  return df['name'].where(df['destination']==dest).dropna()

names = get_passengers_by_destination('Karachi')
print(names)
```

```
#function that returns the name and destination of the passenger who paid the most

def get_highest_paying_passenger():
  return df[['name','destination']].where(df['ticket_price']==max(df['ticket_price'])).dropna()

high_value = get_highest_paying_passenger()
print(high_value)
```

```
#sorted passengers by ticket_price in descending order

df.sort_values(by='ticket_price',ascending=False)
```

```
#top 3 passengers with their name, age, and ticket price.

df[['name', 'age','ticket_price']].sort_values(by='ticket_price',ascending=False).head(3)
```

```
#function that checks passengers that are younger than given age.


def check_age(age):
  print(df['name'].where(df['age']<age).dropna().count())

check_age(25)
```
